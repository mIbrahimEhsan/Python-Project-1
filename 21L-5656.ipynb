Define a procedure called "check_input" that takes an input as a square list of lists (2D Matrix of N x N) and returns a boolen "true" if the input is valid or "false" otherwise, according to the following criteria.

1. Each coloumn must contain numbers from between 1 and N, with no duplicates.
2. Each row must contain numbers from between 1 and N, with no duplicates.

```python

def check_input(array):  # Function To check input values of the 2-D array
    r_c = 0
    c_c = 0
    x = len(array[0])
    for c in range(0,3):  #for checking column wise
        for r in range(1,3):
            if(array[0][c]==array[r][c] or type(array[r][c])==str):
                c_c = c_c + 1
    for i in range(0,x):  #for checking row wise
        for j in range(1,x):
            if(array[r][0]==array[r][c] or type(array[r][c])==str):
                r_c = r_c + 1   
    if(r_c!=0 or c_c!=0): #bool value
        return False
    else:
        return True
```
